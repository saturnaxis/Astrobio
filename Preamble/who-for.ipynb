# Who are these notes for?

This Jupyter Book should be useful for:

- Undergraduates with some *introductory* background in physics

- Anyone else with some astronomy or physics background who is interested!

## Interacting with the Book
The code snippets in the book are all written in Python through Jupyter notebooks. Familiarity with the Python language and its scientific packages is useful, but not required.  There are introductory guides to [Python](https://saturnaxis.github.io/ModernAstro/Preamble/Python-basics.html) and [Markdown](https://saturnaxis.github.io/ModernAstro/Preamble/Markdown-basics.html) in the **Preamble**.

This book is meant to be interactive, where you can open each chapter using [Google Colab](https://research.google.com/colaboratory/) (see below) from the dropdown menu shown with a rocket.  The runtime on Google Colab uses a version of Python that already has many useful packages (e.g., numpy, scipy, pandas, etc.) available with an import statement.  Each cell can be modified and executed, which is only limited by your curiosity.  Also, the text cells can be opened to reveal the source markup and formatting.

## Cloning the Book

All of the source material is publicly available in a GitHub repository [saturnaxis/ModernPhysics](https://github.com/saturnaxis/ModernPhysics).  The open source nature of GitHub [![GitHub](https://img.shields.io/badge/--181717?logo=github&logoColor=ffffff)](https://github.com/) allows you to clone the book to have a local copy.  Note that this book is also meant to be a **living document**, where the content will continue to evolve and collaboration is welcome.
